### Logistic Regression From Scratch

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
def z_func(x, w, b):
    return np.dot(x, w) + b
```

```python
def f_func(x, w, b):
    out = sigmoid(z_func(x, w, b))
    return out
```

```python
def cost_func(X, y, w, b):
    m, n = X.shape
    summation_loss = 0

    for i in range(m):
        loss = (y[i] * np.log(f_func(X[i], w, b)) + (1 - y[i]) * np.log(1 - f_func(X[i], w, b)))
        summation_loss += loss
    total_loss = - (summation_loss) / m

    return total_loss
```

```python
X = np.array([[1, 2],
              [2, 3],
              [3, 4],
              [5, 6],
              [6, 7],
              [7, 8]])

y = np.array([0, 0, 0, 1, 1, 1])

w = [0, 0]

b = 0
```

```python
cost_func(X, y, w, b)
```
