```python
# Entropy Calculation
def entropy_compute(y):
    entropy = 0
    edible_length = 0

    for item in y:
        if(item == 1):
            edible_length += 1
    if(len(y) != 0):
        p1 = edible_length / len(y)
    else:
        p1 = 0
    
    if(p1 != 0 and p1 != 1):
        entropy = -p1 * np.log2(p1) - (1 - p1) * np.log2(1 - p1)
    else:
        entropy = 0
    
    return entropy
```

```python
print("Entropy at root Node:", entropy_compute(y_train))
```
