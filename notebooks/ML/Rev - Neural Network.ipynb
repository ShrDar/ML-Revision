# Neural Network

## Importing Essential Packages

```python
import numpy as np

import tensorflow as tf

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
```

## Building a Neural Network

### Neural Network Implementation from Scratch

```python
# Hard Coding Layers

def sigmoid(z):
    return (1/(1 + np.exp(-z)))

x = np.array([[2, 17], 
              [40, 50]])
x = x.T

# Layer 1
w1_1 = np.array([1, 2])
b1_1 = np.array([-1])
z1_1 = np.matmul(w1_1, x) + b1_1
a1_1 = sigmoid(z1_1)

w1_2 = np.array([-3, 4])
b1_2 = np.array([1])
z1_2 = np.matmul(w1_2, x) + b1_2
a1_2 = sigmoid(z1_2)

w1_3 = np.array([5, -6])
b1_3 = np.array([2])
z1_3 = np.matmul(w1_3, x) + b1_3
a1_3 = sigmoid(z1_3)

a1 = np.array([a1_1, a1_2, a1_3])

#Layer 2 Output Layer

w2_1 = np.array([-7, 8, 9])
b2_1 = np.array([3])
z2_1 = np.matmul(w2_1, a1) + b2_1
a2_1 = sigmoid(z2_1)

a2 = np.array([a2_1])

a1, a2
```

```python
def g(z):
    return (1/(1 + np.exp(-z)))

def dense(a_in, W, b):
    units = W.shape[1]
    a_out = np.zeros(units)
    for j in range(units):
        w = W[:, j]
        z = np.dot(w, a_in) + b[j]
        a_out[j] = g(z)
    return a_out

```

```python
X = np.array([
    [0.1, 0.2, 0.3],
    [0.5, 0.7, 0.2],
    [0.9, 0.1, 0.4],
    [0.3, 0.8, 0.6],
    [0.2, 0.2, 0.9],
    [0.8, 0.5, 0.3],
    [0.6, 0.9, 0.7],
    [0.4, 0.3, 0.8],
    [0.7, 0.4, 0.5],
    [0.2, 0.6, 0.1]
])

# Binary labels
Y = np.array([
    [0],
    [1],
    [1],
    [1],
    [0],
    [1],
    [1],
    [0],
    [1],
    [0]
])
```

```python
# Binary Classification with Sigmoid activaton
model = Sequential([
    Dense(units = 25, activation = "sigmoid"),
    Dense(units = 15, activation = 'sigmoid'),
    Dense(units = 1, activation = "sigmoid")
])

model.compile(loss = BinaryCrossentropy())

model.fit(X, Y, epochs = 1000) #epochs => number of steps in gradient descent
```

```python
# Multiclass Classification with Relu in Hidden and Softmax for output layer

multi_class_model = Sequential([
    Dense(units = 25, activation = "relu"),
    Dense(units = 15, activation = 'relu'),
    Dense(units = 10, activation = 'softmax')
])

multi_class_model.compile(loss = SparseCategoricalCrossentropy())

multi_class_model.fit(X, Y, epochs = 100)
```

#### Improved Implementation (Recommended)

```python
improved_binary_model = Sequential([
    Dense(units = 25, activation = "relu"),
    Dense(units = 15, activation = 'relu'),
    Dense(units = 1, activation = "linear")
])

improved_binary_model.compile(BinaryCrossentropy(from_logits = True))

improved_binary_model.fit(X, Y, epochs = 100)

binary_logits = improved_binary_model(X)

f_x = tf.nn.softmax(binary_logits)
```

```python
improved_softmax_model = Sequential([
    Dense(units = 25, activation = "relu"),
    Dense(units = 15, activation = "relu"),
    Dense(units = 10, activation = "linear")
])

improved_softmax_model.compile(SparseCategoricalCrossentropy(from_logits=True))

improved_softmax_model.fit(X, Y, epochs = 100)

softmax_logits = improved_softmax_model(X)

f_x = tf.nn.softmax(softmax_logits)

# Logits are the numbers which is acquired after passing the inputs 
# through the linear function
# if logits are passed through the activation function and used in the loss function,
# it can cause numerical round off error due to large floating point numbers
# thus to let tensorflow handle this situation, logits are used instead of probabilty values acquired from the 
# activation function
```

```python
# Using Adam Optimizer

adam_model = Sequential([
    Dense(units = 25, activation = 'relu'),
    Dense(units = 15, activation = 'relu'),
    Dense(units = 10, activation = 'linear'),
])

adam_model.compile(optimizer = Adam(learning_rate = 1e-3), loss = SparseCategoricalCrossentropy(from_logits = True))
# automatically adapts the learning rate

adam_model.fit(X, Y, epochs = 100)

adam_logits = adam_model(X)
f_x = tf.nn.softmax(adam_logits)
```
