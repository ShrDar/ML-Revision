## Matplotlib

```python
import matplotlib.pyplot as plt
```

### Matplotlib Basic Intro

```python
# Matplotlib => It's a python library used for data visualization.
# It helps in turining data into charts, bar charts, histograms, scatter plots and so on
```

```python
x = np.array([2023, 2024, 2025, 2026, 2027])
y = np.array([12, 20, 35, 40, 42])

plt.plot(x, y) #if one argument is not given the y will be the default argument and x will be supposed as 0 to the number of values in y with increment of 1 in the x coordinate
plt.show()
```

```python
year = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026])
chelTrophies = np.array([1, 3, 0, 2, 1, 2, 0])

plt.plot(year, chelTrophies)
plt.show()
```

### Plot Customization

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 24, 30, 20])
y2 = np.array([2, 10, 40, 30])
y3 = np.array([20, 50, 80, 120])

line_style = dict(marker = ".", #dict() changes the something = smthvalue to dictionary key value pair
                markersize = 15, # shorthand => ms
                markerfacecolor = "#FB9393", # => mfc
                markeredgecolor = "#000000", # => mec
                linestyle = "dotted",
                linewidth = 2)

plt.plot(x, y1, **line_style, color = "#ED8631") # solid, dashdot, dotted, None (** => unpack to original state)

plt.plot(x, y2, **line_style, color = "#000000")

plt.plot(x, y3, **line_style, color = "#009043")


plt.show()
```

### Labels

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 24, 30, 20])
y2 = np.array([2, 10, 40, 30])
y3 = np.array([20, 50, 80, 120])

line_style = dict(marker = ".", #dict() changes the something = smthvalue to dictionary key value pair
                markersize = 15, # shorthand => ms
                markerfacecolor = "#FB9393", # => mfc
                markeredgecolor = "#000000", # => mec
                linestyle = "solid", # solid, dashdot, dotted, None (** => unpack to original state)
                linewidth = 2)

plt.title("Class Size", fontsize = 15, 
                        family = "Arial",
                        fontweight = "bold",
                        color = "#565656")
plt.xlabel("Year", fontsize = 14, 
                    fontweight = "bold",
                    color = "#565656")
plt.ylabel("No. of Students", fontsize = 14, 
                    fontweight = "bold",
                    color = "#565656")
plt.tick_params(axis = "both", 
                colors = "#3A3A3A")

plt.plot(x, y1, **line_style, color = "#ED8631") 

plt.plot(x, y2, **line_style, color = "#000000")

plt.plot(x, y3, **line_style, color = "#009043")

plt.xticks(x)


plt.show()
```

### Grid Lines

```python
# grid() => helps in making plots easier to read by adding reference lines.

x = np.array([1, 2, 3, 4, 5])
y = np.array([20, 30, 40, 50, 90])


plt.plot(x, y, marker = ".", 
                markersize = 20, 
                markeredgecolor = "#515151", 
                markerfacecolor = "#2A2A2A",
                color = "#000000")

plt.grid(axis="both", 
         visible=None,
         linewidth = 1,
         color = "#000000",
         linestyle = "dotted")

plt.xticks(x)

plt.title("Grid Lines Demo")
plt.xlabel("No. of Guns")
plt.ylabel("No of Deaths")

plt.show()
```

### Bar Charts

```python
# Bar Chart => compares categories of data by representing each category with a bar

categories = ["Grains", "Fruits", "Veggies", "Protein", "Dairy", "Sweets"]
values = np.array([4, 3, 2, 5, 2, 1])

# plt.bar(categories, values, color = "#f3d57b")
plt.barh(categories, values, color = "#f3d57b")


plt.title("Daily Food Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.show()
```

### Pie-Charts

```python
# Pie-Chart => Circular chart divided into slices to show percentages of the total.
# Good for visualizing distribution among categores

categories = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])
colors = ["#9469B5", "#ffacac", "#fff7a1", "#a5baff"]

plt.pie(values, labels=categories, 
                autopct="%1.1f%%",
                colors = colors,
                explode=[0, 0, 0, 0.1],
                shadow = True,
                startangle=70)

plt.title("College Student Category Distribution")
plt.show()
```

### Scatter Graph/Plot

```python
# Scatter Graph => Shows relationship between variables 
# Helps to identify a correlation (+, -, None)
# Example: Study hours and Test Scores

x1 = np.array([0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 9])
y1 = np.array([39, 57, 59, 60, 68, 75, 77, 78, 79, 82, 91])

x2 = np.array([0, 1, 2, 3, 3, 4, 5, 8, 9, 10, 11])
y2 = np.array([10, 20, 42, 50, 55, 66, 75, 79, 88, 90, 92])

plt.scatter(x1, y1, color = "#000000", 
                    alpha = 0.5,
                    s = 100, 
                    edgecolors="000000", 
                    label = "Class A")

plt.scatter(x2, y2, color = "#4AB2B0", 
                    s = 90, 
                    alpha=0.7,
                    edgecolor="#694800",
                    label = "Class B")

plt.title("Scatter Plot of Study Hours Vs Test Scores")
plt.xlabel("Hours")
plt.ylabel("Test Scores")

plt.legend()
plt.show()
```

### Histograms

```python
# Histogram => Visual Representation of the distribution of quantitative data
# They group values into bins (intervals)
# and counts how many fall in each range


scores = np.random.normal(loc = 80, scale = 10, size = 100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins = 10,
                color = "lightgreen",
                edgecolor = "#000000")

plt.title("Scores")
plt.xlabel("Score")
plt.ylabel("# of Students")

plt.show()
```

### Subplots

```python
# Figure => The Entire Canvas
# Ax = A single plot (subplot)

x = np.array([1, 2, 3, 4, 5])

figure, axes = plt.subplots(2,2)


axes[0, 0].plot(x, 2*x + 3, color = "#444444A3")
axes[0, 0].set_title("Linear Equation")

axes[0, 1].plot(x, (7*x**2)/2, color = "#FFBE4E")
axes[0, 1].set_title("Quadratic Equation")

axes[1, 0].plot(x, (x**3) + 9, color = "#ff5bc0")
axes[1, 0].set_title("Cubic Equation")

axes[1, 1].plot(x, (x**4) - 20, color = "#63fe6b")
axes[1, 1].set_title("Quintic Equation")


plt.tight_layout()
plt.show()
```

### Matplotlib and Pandas

```python
df = pd.read_csv("data/pokemon.csv")
```

```python
# Alternative Method of finding the number occurance of each category

pokemonNames = (df['Type1'].unique())
pokeSums = []

for pokemon in pokemonNames:
    sum = 0
    for name in df['Type1']:
        if(pokemon == name):
            sum += 1
    pokeSums.append({pokemon: sum})

pokeSums
```

```python
print(df['Type1'].value_counts())
```

```python
type_count = df["Type1"].value_counts(ascending=True)


print(type_count.index) # list of indexes
print(type_count.values) # list of values
```

```python
plt.barh(type_count.index, 
         type_count.values, 
         color = "#777777",
         edgecolor = "#000000")

plt.title("# of Pokemon by Primary Type")
plt.xlabel("No. of Pokemons")
plt.ylabel("Type")

plt.tight_layout()
plt.show()
```
