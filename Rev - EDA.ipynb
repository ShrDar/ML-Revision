### Missing Values Check and Duplicated Rows Check

```python
subsetted_df.isna().sum()
```

```python
# Checking for Duplicated Rows
subsetted_df.loc[subsetted_df.duplicated()] 
```

```python
subsetted_df.loc[subsetted_df.duplicated(subset = ["Coaster_name"])]
```

```python
# Checking an example duplicate
subsetted_df.query("Coaster_name == 'Crystal Beach Cyclone'")
```

```python
subsetted_df[subsetted_df['Coaster_name'] == "Crystal Beach Cyclone"]
```

```python
# ~ is the inverse of the result

subsetted_df = subsetted_df.loc[~subsetted_df.duplicated(subset = ['Coaster_name', "Opening_Date"])]
```

```python
# After removing the duplicated rows, the indexes may not be serial as there can be a jump in index
subsetted_df
```

```python
subsetted_df = subsetted_df.reset_index(drop = True)
```

## Ask a Question about the Data

> Try to answer the questioni you have about the data using a plot or statistics

> - What are the locations with the fastest roller coasters? (minimum of 10)

```python
question1_df = subsetted_df[subsetted_df['Location'] != "Other"] \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values('mean')

ax = sns.barplot(question1_df['mean'], orient = "v")

ax.set_title("Top 10 Locations with the fasters roller coasters")
ax.set_xlabel("Speed Mean")
ax.set_ylabel("Location")

plt.xticks(rotation = 60)
plt.show()
```
